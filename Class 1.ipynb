# Problem 1
## Anagram Check
Create a function called `anagram_check` that takes two strings as parameters and returns True if these are anagrams of each other, and False otherwise.

### Criteria
Two strings are said to be anagrams of each other if all the crietira below are met:
1. They must contain the same amount of letters.
2. They must use the same letters the same amount of times.

### Examples
```
heart and earth
scar and cars
rats and star
aabb and baba
```

## Pseudocode for problem 1 (one possible solution)
```
define function "anagram_check" wich recieves string1 and string2 as paremeters:
    string1 = string1 without spaces and in lower case
    string2= string2 without spaces and in lower case
    if string1 and string2 are not equal in length:
        return False
    dict1 = new dictionary
    dict2 = new dictionary
    for each "number" in the range between 0 and the length of string1:
        if the character at the position "number" in string1 is dict1:
            add 1 to the value associated with that character # this represents how many times we've seen that character
        else:
            create a new dictionary entry with the character at position "number" setting its value to 1
        if the character at the position "number" in string2 is dict2:
            add 1 to the value associated with that character # this represents how many times we've seen that character
        else:
            create a new dictionary entry with the character at position "number" setting its value to 1
    if dict1 is equal to dict2:
        return True
    otherwise:
        return False
```

````python
def anagram_check(string1, string2):
    str1 = string1.replace(" ", "").lower()
    str2 = string2.replace(" ", "").lower()

    if len(str1) != len(str2):
        return False

    dict1 = {}
    dict2 = {}

    for x in range(len(str1)):
        if str1[x] not in dict1:
            dict1[str1[x]] = 1
        elif str1[x] in dict1:
            dict1[str1[x]] = dict1.get(str1[x]) + 1
        if str2[x] not in dict2:
            dict2[str2[x]] = 1
        elif str2[x] in dict2:
            dict2[str2[x]] = dict2.get(str2[x]) + 1
            
    print(string1, dict1)
    print(string2, dict2)

    if dict1 == dict2:
        return True
    else:
        return False

print(anagram_check("heart", "earth"))
print(anagram_check("scar", "cars"))
print(anagram_check("rats", "scar"))
print(anagram_check("aabb", "baba"))
print(anagram_check("I am lord Voldemort", "Tom Marvolo Riddle"))
print(anagram_check("AbC", "xYz"))
print(anagram_check("Not equal", "Equal"))
````
